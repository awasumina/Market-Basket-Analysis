```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
from google.colab import files
uploaded = files.upload()
df = pd.read_excel ('Online Retail.xlsx')
df.head()
```

```python
#cleaning data
# Removing extra spaces in the Description column.
# Dropping rows with missing invoice numbers.
# Ensuring all invoice numbers are treated as strings.
# Excluding any transactions that were canceled.
   
df['Description'] = df['Description'].str.strip()
df.dropna(axis=0, subset=['InvoiceNo'], inplace=True)
df['InvoiceNo']=df['InvoiceNo'].astype('str')
df=df[~df['InvoiceNo'].str.contains('C')]
df
```

```python
# The code creates a table (`basket`) that shows the quantity of each item bought in each transaction made in France. Each row is a transaction (`InvoiceNo`), and each column is an item (`Description`). If an item wasn't bought in a transaction, its quantity is set to `0`.
basket=(df[df['Country']=="France"]
.groupby(['InvoiceNo', 'Description'])['Quantity']
.sum().unstack().reset_index().fillna(0)
.set_index('InvoiceNo'))
basket
```

```python
# The code converts the quantities in the `basket` DataFrame into binary values, where `0` indicates no purchase and `1` indicates a purchase. It then removes the `POSTAGE` column from the resulting DataFrame (`basket_sets`) and displays the modified data.

def encode_units (x):
  if x <= 0:
    return 0
  if x >= 1:
    return 1
basket_sets= basket.applymap(encode_units)
basket_sets.drop('POSTAGE', inplace=True, axis=1)
basket_sets

```

```python
# The code identifies frequent itemsets in the transactions, generates association rules based on the lift metric, and displays the top results.


frequent_itemsets = apriori (basket_sets, min_support=0.07, use_colnames=True)
rules= association_rules (frequent_itemsets, metric="lift", min_threshold=1)
rules.head()
```

```python
# The code filters the `rules` DataFrame to find association rules where the lift is at least 6 and the confidence is at least 0.8. This identifies strong and reliable rules based on these metrics.

rules[(rules['lift']>=6) &(rules['confidence']>=0.8)]
```
